```python
final_state=pd.read_csv('final_state_data.csv')
import matplotlib.pyplot as plt
%matplotlib inline
final_state1=final_state.loc[~final_state['state'].isin(['New York','California','Florida','Texas']),:]
```

```python

fig=plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.scatter('gdp','net_change',data=final_state1,color='orange')
ax1.title.set_text('Gdp vs Net change')
ax1.title.set_fontsize(15)
ax1.set_xlabel('Gross domestic product')
ax1.set_ylabel('Net Change')
ax2.scatter('person consumption expenditure','net_change',data=final_state1,color='orange')
ax2.title.set_text('Personal income vs Net change')
ax2.set_xlabel('personal income')
ax2.title.set_fontsize(15)
ax3.scatter('total_engi','net_change',data=final_state1,color='orange')
ax3.title.set_text('Graduate Students in Engineer vs Net Change')
ax3.set_xlabel('Ggraduate students')
ax3.set_ylabel('total expanding number of establishments')
ax3.title.set_fontsize(15)
ax4.scatter('member of unions','net_change',data=final_state1,color='orange')
ax4.title.set_text('Total Union Members vs Net_change')
ax4.set_xlabel('Total union members')
ax4.set_xlim()
ax4.title.set_fontsize(15)


```

```python

fig=plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.scatter('gdp','total_expand',data=final_state1,color='orange')
ax1.title.set_text('Gdp vs Expanding Number')
ax1.title.set_fontsize(15)
ax1.set_xlabel('Gross domestic product')
ax1.set_ylabel('total expanding number of establishments')
ax2.scatter('person consumption expenditure','total_expand',data=final_state1,color='orange')
ax2.title.set_text('Personal Income vs Expanding Number')
ax2.set_xlabel('personal income')
ax2.title.set_fontsize(15)
ax3.scatter('total_engi','total_expand',data=final_state1,color='orange')
ax3.title.set_text('Graduate Students vs Expanding Number')
ax3.set_xlabel('Ggraduate students')
ax3.set_ylabel('total expanding number of establishments')
ax3.title.set_fontsize(15)
ax4.scatter('member of unions','total_expand',data=final_state1,color='orange')
ax4.title.set_text('Total Union Members vsc Expanding Number')
ax4.set_xlabel('total union members')
ax4.title.set_fontsize(15)


```
