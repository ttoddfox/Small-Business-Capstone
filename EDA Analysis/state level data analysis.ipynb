```python
eda =state_merge_data.copy()
eda['close_firms'] = -eda['ending_establishments']
```

```python
sns.set()
f, ax = plt.subplots(figsize=(10, 12))

eda.sort_values(['opening_establish'], ascending=False, inplace=True)
sns.set_color_codes("muted")
g1 =sns.barplot(x='opening_establish', y='state', data=eda, color="g",label ='# Opening Firms')
g2 =sns.barplot(x='close_firms', y='state', data=eda, color="r",label ='# Closing Firms')
#g1.despine(left=True).set_xlabels('# Opening Firms')
#g2.despine(left=True).set_xlabels('# Closing Firms')
#sns.despine(left=True, bottom=True)
#plt.show()
```

```python
sns.set()

#depentent variables
state_merge_data.sort_values(['opening_establish'], ascending=False, inplace=True)

f, ax = plt.subplots(nrows=1, ncols=2,figsize=(14, 15))

#add col 1 -- opening_establish
sns.set_color_codes("muted")
sns.barplot(x='opening_establish', y='state', data =state_merge_data,color="g",ax=ax[0])
ax[0].set(ylabel="",xlabel="#Total Firms Opening by State")
sns.despine(left=True, bottom=True)

#add col 2 -- ending_establishments 
state_merge_data.sort_values(['ending_establishments'], ascending=False, inplace=True)
sns.set_color_codes("muted")
sns.barplot(x='ending_establishments', y='state', data=state_merge_data,color="r",ax=ax[1])
ax[1].set(ylabel="",xlabel="#Total Firms Bankrupty by State")
sns.despine(left=True, bottom=True)

#add col 3 -- total_expanding 
#state_merge_data.sort_values(['total_expanding'], ascending=False, inplace=True)
#sns.set_color_codes("muted")
#sns.barplot(x='total_expanding', y='state', data=state_merge_data,color="b",ax=ax[2])
#ax[2].set(ylabel="",xlabel="#Total Firms Expanding by State")
#sns.despine(left=True, bottom=True)
```

```python
sns.set()

#depentent variables
state_merge_data.sort_values(['opening_establish'], ascending=False, inplace=True)

f, ax = plt.subplots(nrows=1, ncols=2,figsize=(14, 15))

#add col 1 -- opening_establish
sns.set_color_codes("muted")
sns.barplot(x='opening_establish', y='state', data =state_merge_data,color="b",ax=ax[0])
ax[0].set(ylabel="",xlabel="#Total Firms Opening by State")
sns.despine(left=True, bottom=True)

#add col 2 -- ending_establishments 
state_merge_data.sort_values(['us_population'], ascending=False, inplace=True)
sns.set_color_codes("muted")
sns.barplot(x='us_population', y='state', data=state_merge_data,color="b",ax=ax[1])
ax[1].set(ylabel="",xlabel="#Population by State")
sns.despine(left=True, bottom=True)
```

```python
f, ax = plt.subplots(figsize=(7, 15))

state_merge_data.sort_values(['net_change'], ascending=False, inplace=True)
sns.set_color_codes("muted")
sns.barplot(x='net_change', y='state', data=state_merge_data,color='g')
ax.set(ylabel="",xlabel="#Change of Number of Establishments by State")
sns.despine(left=True, bottom=True)
```

```python
eda3_1 =eda2.copy()
```

```python
eda3 =eda3_1.fillna(0)
```

```python
#scaler = MinMaxScaler()
#NUM_COLUMNS = list(eda3.select_dtypes(exclude=['object']).columns)
#eda3[NUM_COLUMNS] = scaler.fit_transform(eda3[NUM_COLUMNS])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
NUM_COLUMNS = list(eda3.select_dtypes(exclude=['object']).columns)
eda3[NUM_COLUMNS] = scaler.fit_transform(eda3[NUM_COLUMNS])
```

```python
eda3.loc[eda3['state'].isin(['Maine', 'Rhode Island', 'Vermont', 'Connecticut', 'New Hampshire' ,'Massachusetts']), 
          'state_part'] = 'New England Region'
eda3.loc[eda3['state'].isin(['New York', 'New Jersey' ,'Pennsylvania']), 
          'state_part'] = 'Mid-Atlantic Region' 
eda3.loc[eda3['state'].isin(['Virginia', 'West Virginia', 'Kentucky', 'Delaware', 'Maryland', 'North Carolina',
                              'South Carolina','Tennessee', 'Arkansas', 'Louisiana', 'Florida', 'Georgia', 
                              'Alabama','Mississippi','District of Columbia']), 
          'state_part'] = 'Southern Region' 
eda3.loc[eda3['state'].isin(['Michigan', 'North Dakota', 'Iowa', 'Minnesota', 'Kansas', 'Nebraska', 
                              'Ohio', 'Indiana', 'Illinois', 'Wisconsin','Missouri','South Dakota']), 
          'state_part'] = 'Mid-West Region' 
eda3.loc[eda3['state'].isin(['Texas', 'Arizona', 'New Mexico','Oklahoma']), 
          'state_part'] = 'South-West Region' 
eda3.loc[eda3['state'].isin(['Montana', 'Idaho', 'Colorado', 'Utah', 'Wyoming','Nevada']), 
          'state_part'] = 'Rocky Mountains' 
eda3.loc[eda3['state'].isin(['California', 'Oregon','Washington','Hawaii','Alaska']), 
          'state_part'] = 'Pacific Coastal Region' 
```

```python
from pandas.plotting import parallel_coordinates
fig = plt.figure(figsize=(20, 10))

title = fig.suptitle("Feature Distribution by US Region", fontsize=25)
fig.subplots_adjust(top=0.93)

new_cols = ['state_part','total_engi', 'temporary_visa_engi', 'totals_science',
       'temporary_visa_science', 'total_area_sqm', 'land_area',
       'water_area', 'total_employed', 'member of unions', 'represented by unions',
       'sales_tax', 'us_population',#'irp','RPP', 
            'gdp', 'personal income', 'person consumption expenditure',
       'opening_establish','ending_establishments','total_expanding','net_change']
pc = parallel_coordinates(eda3[new_cols],'state_part',color=('#F68E8E','#D38EF6','#2080D4',
                                                             '#20C8D4','#20D474','#BCD420','#F30C0C'))
plt.xticks(rotation=50,size =17)
plt.yticks(size = 20)
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left",prop={'size': 20})
```
