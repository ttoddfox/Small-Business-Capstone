# 1.Load & Prepare Data

```python
def get_quarter_info_target(x):
    if x == "March":
        return 1
    elif x == "June":
        return 2
    elif x == "September":
        return 3
    else:
        return 4
```

**Load & Merge & Prepare Final Data**

```python
final_data = pd.DataFrame()
state_name_list = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California',
                   'Colorado', 'Connecticut', 'Delaware', 'Florida', 'Georgia', 
                   'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'New_York']
for state in state_name_list:
    # Sentiment from tweets
    state_name = str(state).replace("_", " ")
    state_sentiment = pd.read_csv("sentiment_from_tweets/%s_sentiment.csv" % state, index_col=0)
    # Target: up&down in SME
    target = pd.read_csv("target.csv")
    target.year = target.year.apply(lambda x: int(x))
    target["quarter"] = target.month.apply(lambda x: get_quarter_info_target(x))
    target = target.sort_values(["year", "quarter"])
    target["timestamp"] = target.year.apply(lambda x: str(int(x)) + "_") \
                                        + target.quarter.apply(lambda x: str(int(x)))
    target = target[(target.year > 2011) & (target.year <= 2016) & (target.state == state_name)].reset_index(drop=True)
    target = target.drop_duplicates().reset_index(drop=True)
    # Merge
    temp_data = state_sentiment.copy().reset_index(drop=True)
    temp_data = temp_data.merge(target, on="timestamp")
    final_data = final_data.append(temp_data, ignore_index=True)

final_data = final_data.drop(["quarter_x", "year_x"], axis=1)
final_data = final_data[['state', 'year_y','quarter_y', 'timestamp', 'month',
                         'negative', 'neutral', 'positive', 'contract_establish',
                         'end_establish', 'expand_establish', 'net_change',
                         'open_establish', 'total_contract', 'total_expand', ]]
final_data.columns = ['state', 'year', 'quarter', 'timestamp', 'month',
                         'negative', 'neutral', 'positive', 'contract_establish',
                         'end_establish', 'expand_establish', 'net_change',
                         'open_establish', 'total_contract', 'total_expand', ]
```

**Select columns of interest**
* To make this analysis simple, we will take total_contract & total_expand for the target values of interest.

```python
final_data = final_data[['state', 'year', 'quarter', 'timestamp', 'month',
                         'negative', 'neutral', 'positive', 'total_contract', 'total_expand', 'net_change']]
```

## 3-3.Correlation Heapmap per state
* To filtering out cases with clear relationship

```python
state_corr = pd.DataFrame()
for state in final_data.state.unique():
    temp = final_data[final_data.state == state].reset_index(drop=True)
    pos_vs_cont = temp["positive"].corr(temp["total_contract"])
    pos_vs_exp = temp["positive"].corr(temp["total_expand"])
    neg_vs_cont = temp["negative"].corr(temp["total_contract"])
    neg_vs_exp = temp["negative"].corr(temp["total_expand"])
    temp_dict = {"state": state, "pos_vs_cont": pos_vs_cont, "pos_vs_exp":pos_vs_exp, 
                 "neg_vs_cont":neg_vs_cont, "neg_vs_exp":neg_vs_exp}
    state_corr = state_corr.append(temp_dict, ignore_index=True)
```

```python
plt.figure(figsize=(10,10));
sns.heatmap(state_corr.drop("state", axis=1), yticklabels=list(state_corr.state));
plt.title("Correlation Heatmap", fontsize=20);
```

* Followings are cases with relatively clear linear relationship: abs(correlation) > 0.6
    * California: Negative - Total Contract
    * Idaho: Positive - Total Expand
    * New York: Negative - Total Contract
